```python
import pandas as pd
import numpy as np
```

```python
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
```

```python
akas_url ='https://datasets.imdbws.com/title.akas.tsv.gz'
```

```python
ratings_url ='https://datasets.imdbws.com/title.ratings.tsv.gz'
```

```python
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)
```

```python
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)
```

```python
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

```python
basics
```

```python
ratings
```

```python
akas
```

```python
basics.replace({'\\N':np.nan}, inplace=True)
```

```python
basics.isnull().sum()
```

```python
basics
```

```python
basics = basics[basics['runtimeMinutes'].notna()]
```

```python
basics
```

```python
basics['genres'].isnull().sum()
```

```python
basics = basics[basics['genres'].notna()]
```

```python
basics['genres'].isnull().sum()
```

```python
basics = basics[basics['titleType']=='movie']
```

```python
basics = basics [ ( basics['startYear'] < '2022' ) & ( basics['startYear'] > '2000' ) ]
```

```python
basics
```

```python
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
```

```python
(basics['genres']=='documentary').sum()
```

```python
akas = akas[akas['region']=='US']
```

```python
akas
```

```python
akas.replace({'\\N':np.nan}, inplace=True)
```

```python
keepers =basics['tconst'].isin(akas['titleId'])
keepers
```

```python
basics = basics[keepers]                                                                                               
basics
```

```python
ratings.isnull().sum()
```

```python
keepers1 =ratings['tconst'].isin(akas['titleId'])
keepers1
```

```python
ratings = ratings[keepers1]                                                                                               
ratings
```
